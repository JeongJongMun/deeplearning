```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_positional_encoding(max_len, d_model):
    position = np.arange(max_len)[:, np.newaxis]
    div_term = np.exp(np.arange(0, d_model, 2) * -(np.log(10000.0) / d_model))
    
    pos_enc = np.zeros((max_len, d_model))
    pos_enc[:, 0::2] = np.sin(position * div_term)
    pos_enc[:, 1::2] = np.cos(position * div_term)
    
    return pos_enc
```

```python
pos_enc = get_positional_encoding(50, 16)
print(pos_enc[0])
str = ' '.join([f'{value:.2f}' for value in pos_enc[1]])
print(f'[{str}]')
```

```python
# Set the parameters
max_len = 100
d_model = 16  # or 512 as in the original paper

pos_enc = get_positional_encoding(max_len, d_model)

# Visualization
plt.figure(figsize=(12,8))
plt.pcolormesh(pos_enc, cmap='viridis')
plt.xlabel('Embedding Dimensions')
plt.ylabel('Token Position')
plt.colorbar()
plt.title('Positional Encoding')
plt.show()
```
