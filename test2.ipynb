```python
import torch

class sampleClass(torch.nn.Module):
	def __init__(self) -> None:
		super().__init__()
		self.linear1 = torch.nn.Linear(20, 40)
		self.relu = torch.nn.ReLU()
		self.linear2 = torch.nn.Linear(40, 2)
		self.ff = torch.nn.Sequential(
		self.linear1,
		self.relu,
		self.linear2
	)
	def forward(self, x):
		return self.ff(x)

model = sampleClass()
model.train()

x = torch.randn(20)
target = torch.tensor([1., -1.0])
y = model(x)
loss = torch.square(target-y).mean(dim=-1)
loss.backward()
for name, param in model.named_parameters():
	print(name, param.grad)
```

```python
a = torch.tensor([3.0, 2.0], requires_grad=True)
b = torch.tensor(4.0, requires_grad=True)

y = a * b
z = torch.sum(y)
z.backward()

print(z)
print(a.grad)
print(b.grad)
```
